Task 2

```python
text = "Hello, world!"
if text.endswith("!"):
    print("The text ends with an exclamation mark.")

text2 = "Hello, world!"
encoded_text = text2.encode(encoding='utf-8')
print(encoded_text)

text3 = "Hello, World!"
casefolded_text = text3.casefold()
print(casefolded_text)
```
